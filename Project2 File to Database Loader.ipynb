```python
from sqlalchemy import create_engine

#this created connection between table and database
username = 'root'
password = 'khushi' 
port = '3306'
database = 'project2'
engine = create_engine(f'mysql+pymysql://{username}:{password}@localhost:{port}/{database}')
```

<h3><b><u>Once used if_exist then remove it or else it will perform operation as many time as you run it</h3>

```python
import pandas as pd

#to read csv fileand append it to the table
#here first argument of to_sql is tabe name which we create and if it exist then it append the values in it

df = pd.read_csv('sample.csv')

#if I run this command again and again it will append values again and again
#df.to_sql('users', con=engine, if_exists='append', index=False)
df.to_sql('users', con=engine, index=False)
```

```python
#here we just run simple select command of sql to check the values in sql
df_from_db = pd.read_sql("select * from users", con=engine)
print(df_from_db)
```

```python
#Jupyter Notebook Magic Commands are special commands that start with % or %% and give extra functionality inside a Jupyter Notebook.
#They are called “magic” because they add powerful features that are not part of normal Python.

#%%sql is a Jupyter Notebook magic command provided by the ipython-sql extension. It allows you to write and run SQL queries directly inside a notebook cell.
%%sql 

create table if not exists emp(
    id int primary key,
    name varchar(100),
    age int,
    email text
);
```

```python
df2 = pd.DataFrame(
    {
        'id' : [101,102],
        'name' : ['John','Bob'],
        'age' : [40,30],
        'email' : ['john@mail.com','bob@email.com']
    }
)

df2.to_sql('emp', con=engine, index=False)
```

```python
#divide huge amount of data into small chunks and then import

chunk_size=3
for chunk in pd.read_csv('sample.csv',chunksize = chunk_size):
    chunk.to_sql('users',con=engine, if_exists = 'append', index = False)
```

```python
import os
data_dir = 'data/'

for file in os.listdir(data_dir):
    if file.endswith('.csv'):
        table_name = os.path.splitext(file)[0]
        print(f'loading {file} into table {table_name}')
        df = pd.read_csv(os.path.join(data_dir, file))
        df.to_sql(table_name,con = engine , if_exists = 'replace', index = False)
```
